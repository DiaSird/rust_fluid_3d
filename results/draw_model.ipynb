```python
import pandas as pd
import matplotlib.pyplot as plt

# title
title = 'Fluid-Box'
# title = 'NACA2412-Wing'

# Read data
file_path = "model_particles.csv"
data = pd.read_csv(file_path)
data.columns = data.columns.str.strip()

# x, y, z
x = data['x']
# y = data['y']
# z = data['z']
z = data['y']
y = data['z']

# 3D plot
fig = plt.figure(figsize=(12, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c=z, cmap='jet', marker='.')

# axis label
ax.set_xlabel('x [m]')
ax.set_ylabel('y [m]')
ax.set_zlabel('z [m]')

# title
ax.set_title(title)

# show
plt.savefig(f"{title}.png")
plt.show()
```
